```python
import os
from itertools import product
```

```python
maps = [('sorting_map.grid', 'SORTING'), ('kiva.map', 'KIVA')]
methods = ['WHCA', 'ECBS', 'PBS']
windows = [(2, 5), (5, 10), (5, 30), (5, 40), (10, 20), (20, 40), (15, 20)]
command_template = '../lifelong -m ../maps/{_map} -k 100 --scenario={scenario} --simulation_window={h} --planning_window={w} --solver={solver} --seed=0 -o test_data/'
```

```python
results = {_map[0]: {method: {window: (0, 0) for window in windows} for method in methods} for _map in maps}
for _map, method, window_params in product(maps, methods, windows):
    result = os.system(command_template.format(
        _map=_map[0],
        scenario=_map[1],
        h=window_params[0],
        w=window_params[1], 
        solver=method))
    print(result)
    df = group_tasks_by_timestamps(read_tasks('../test_data/\\tasks.txt'))
    results[_map[0]][method][window_params] = (sum(d[1] for d in df)/len(df), sum(d[1] for d in df))
print(results)
```
